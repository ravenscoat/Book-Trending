# Trending Based System

We are making a trend system based on mean average

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
rating = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
rating.rename(columns={'Book-Rating':'rating'},inplace=True)
rating
```

```python
average = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
average.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
average
```

```python
trending=rating.merge(average,on='Book-Title')
```

```python
trending
```

```python
trending = trending[trending['rating']>=200].sort_values('avg_rating',ascending=False).head(50)
```

```python
trending.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','rating','avg_rating']]
```

```python
TRENDING=trending.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','rating','avg_rating']]
```

```python
import pickle

# Dump the TRENDING dataframe into a pickle file
with open('trending_books.pkl', 'wb') as file:
    pickle.dump(TRENDING, file)
```
